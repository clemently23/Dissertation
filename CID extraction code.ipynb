```python
import numpy as np
from urllib.request import urlopen
import cv2
import os
import json
import urllib.request
```

```python
def downloadImage(url, feature_type):
    try:
        print("Downloading %s" % (url))
        
        image_name = str(url).split('/')[-1] # Extract the photo id from the image url
        
        
        #defining header
        header= {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' 
        'AppleWebKit/537.11 (KHTML, like Gecko) '
        'Chrome/23.0.1271.64 Safari/537.11',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
        'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
        'Accept-Encoding': 'none',
        'Accept-Language': 'en-US,en;q=0.8',
        'Connection': 'keep-alive'}

        #the URL where you are requesting at
        req = urllib.request.Request(url=url, headers=header) 
        resp = urlopen(req)

        image = np.asarray(bytearray(resp.read()), dtype="uint8")
        image = cv2.imdecode(image, cv2.IMREAD_COLOR)
        
        path = os.getcwd() + '/Images/' + feature_type
        
        cv2.imwrite(os.path.join(path , image_name), image)
    
    except Exception as error:
        print(error)
```

```python
# Cycle Parking

url = "https://cycling.data.tfl.gov.uk/CyclingInfrastructure/data/points/cycle_parking.json"
response = urllib.request.urlopen(url)
cycle_parking = json.loads(response.read())

# DOWNLOAD images to folder in desktop 
# Change the parameters in range() to determine how many photos
# Change the second parameter in downloadImage() to save it to different folder

for i in range(237,240):
  url_photo = cycle_parking['features'][i]['properties']['PHOTO1_URL']
  downloadImage(url_photo, 'cycle_parking')
```
